```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from apyori import apriori
```

```python
data = pd.read_csv('book.csv', header=None)
```

```python
data.head()
```

```python
data.shape
```

```python
records = []
for i in range(0, 250):
    records.append([str(data.values[i,j]) for j in range(0, 5)])
```

```python
association_rules = apriori(records, min_support=0.0045, min_confidence=0.2, min_length=2)
association_results = list(association_rules)
```

```python
print(len(association_results))
```

```python
print(association_results[0])
```

```python
for item in association_results:
    # No need to extract pair, as item is already the RelationRecord you want to work with.
    # Extract support directly from the item.
    support = item.support
    # Assuming there's only one ordered statistic per RelationRecord, extract it.
    ordered_statistic = item.ordered_statistics[0]
    base_items = ordered_statistic.items_base
    added_items = ordered_statistic.items_add
    confidence = ordered_statistic.confidence
    lift = ordered_statistic.lift
    
    # Print the rule (base_items -> added_items) and its support, confidence, and lift.
    print(f"Rule: {base_items} -> {added_items}")
    print(f"Support: {support}")
    print(f"Confidence: {confidence}")
    print(f"Lift: {lift}")
    print("==========")

    
```
